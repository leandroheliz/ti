```python
import requests
import pandas as pd
from google.colab import files
from IPython.display import display, HTML, display_html
from datetime import datetime
import ipywidgets as widgets
```

```python
LH_API_ENDPOINT = "https://www.googleapis.com/pagespeedonline/v5/runPagespeed"
API_KEY = 'AIzaSyC1GDGsPsYmv7H-tbbgbEEMhCPLotTdWIk'
```

```python
def colour_cell(val, metric):
    try:
        val = float(val)
    except ValueError:
        return ''

    if metric == "LCP":
        if val <= 2.5:
            return 'background-color: green'
        elif val <= 4:
            return 'background-color: yellow'
        else:
            return 'background-color: red'
    elif metric == "CLS":
        if val <= 0.1:
            return 'background-color: green'
        elif val <= 0.25:
            return 'background-color: yellow'
        else:
            return 'background-color: red'
    elif metric == "TBT":
        if val <= 300:
            return 'background-color: green'
        elif val <= 600:
            return 'background-color: yellow'
        else:
            return 'background-color: red'

uploaded = files.upload()
filename = list(uploaded.keys())[0]

if API_KEY == 'AIzaSyC1GDGsPsYmv7H-tbbgbEEMhCPLotTdWIk':
    API_KEY = input("Porfavor introduce tu API Key de PageSpeed Insights: ")

def test_performance(url, df_results):
    params = {
        'url': url,
        'strategy': 'desktop',
        'category': ['performance'],
        'key': API_KEY
    }

    response = requests.get(LH_API_ENDPOINT, params=params)

    if response.status_code == 200:
        data = response.json()
        lcp = float(data['lighthouseResult']['audits']['largest-contentful-paint']['numericValue']) / 1000
        cls = float(data['lighthouseResult']['audits']['cumulative-layout-shift']['numericValue'])
        tbt = float(data['lighthouseResult']['audits']['total-blocking-time']['numericValue'])

        new_row = pd.DataFrame({'URL': [url], 'LCP (s)': [lcp], 'CLS': [cls], 'TBT (ms)': [tbt]})
        df_results = pd.concat([df_results, new_row], ignore_index=True)

    else:
        print(f"Error for {url}: {response.status_code}")

    return df_results

df_urls = pd.read_csv(filename)
df_results = pd.DataFrame(columns=['URL', 'LCP (s)', 'CLS', 'TBT (ms)'])
current_datetime = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
header_row = pd.DataFrame({'URL': [f"Core Web Vitals test run on {current_datetime}"], 'LCP (s)': [""], 'CLS': [""], 'TBT (ms)': [""]})
df_results = pd.concat([df_results, header_row], ignore_index=True)

total_urls = len(df_urls['URL'])
for idx, url in enumerate(df_urls['URL'], start=1):
    print(f"Processing URL {idx} of {total_urls}...")
    df_results = test_performance(url, df_results)

styled_df = df_results.style.applymap(lambda x: colour_cell(x, 'LCP'), subset=pd.IndexSlice[:, ['LCP (s)']])
styled_df = styled_df.applymap(lambda x: colour_cell(x, 'CLS'), subset=pd.IndexSlice[:, ['CLS']])
styled_df = styled_df.applymap(lambda x: colour_cell(x, 'TBT'), subset=pd.IndexSlice[:, ['TBT (ms)']])
display(styled_df)

csv_filename = "resultado_de_performance.csv"
df_results.to_csv(csv_filename, index=False)

def download_csv(b):
    files.download(csv_filename)

button = widgets.Button(description="Descargar Achivo CSV")
button.on_click(download_csv)
display(button)
```
